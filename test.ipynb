```python
import torchvision
import torch
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open("./screenshot.jpg")
```

```python
tsfm = torchvision.transforms.Compose([
    torchvision.transforms.Resize((96, 128)),
    torchvision.transforms.ToTensor()
])
```

```python
tsfm(image).shape
```

```python
plt.matshow(tsfm(image).permute(1,2,0))
```

```python
model = torch.torch.nn.Sequential(
    torch.nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1),  # 1 input chatorch.nnel -> 16 output chatorch.nnels
    torch.nn.ReLU(),
    torch.nn.MaxPool2d(kernel_size=2, stride=2),  # Downsample by 2x (128x128 -> 64x64)

    torch.nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1),  # 16 -> 32 chatorch.nnels
    torch.nn.ReLU(),
    torch.nn.MaxPool2d(kernel_size=2, stride=2),  # Downsample by 2x (64x64 -> 32x32)

    torch.nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),  # 32 -> 64 chatorch.nnels
    torch.nn.ReLU(),
    torch.nn.MaxPool2d(kernel_size=2, stride=2)  # Downsample by 2x (32x32 -> 16x16)
)

```

```python
img = tsfm(image)
```
